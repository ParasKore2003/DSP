```python
#libraries required
import numpy as np
import matplotlib.pyplot as plt
from collections import deque
```

```python

x = [1, 3, -2, 1, 2, -1, 4, 4, 2]
y= [2, -1, 4, 1, -2, 3, 0, 0, 0]
```

```python
#1
#Circular Convolution in Time domain
#x*y = summation(x[i]*y[n-i])modN
z_t = []
y_change = np.flip(y)
for i in range(len(x)):
    sum = 0
    y1 = deque(y_change)
    y1.rotate(i+1)
    for j in range(len(x)):
           sum+=(x[j]*y1[j])
    z_t.append(sum)
plt.stem(z_t)
```

```python
#2
#Circular Convolution in Frequency domain
X = np.fft.fft(x)
Y = np.fft.fft(y)
Z = X*Y
z = np.fft.ifft(Z)
plt.stem(z)
```

```python
#libraries required
import numpy as np
import matplotlib.pyplot as plt
from collections import deque
import time
```

```python
#input
x = [1, 3, -2, 1, 2, -1, 4, 4, 2]
y = [2, -1, 4, 1, -2, 3, 0, 0, 0]

#INput without zero padding
x_wp = [1, 3, -2, 1, 2, -1, 4, 4, 2]
y_wp = [2, -1, 4, 1, -2, 3]

XX = np.hstack((x_wp, np.zeros(len(y_wp)+ 1)))
YY = np.hstack((y_wp, np.zeros(len(x_wp)+ 1)))
```

```python
def time_domain(A,B):
    zt = []
    B1 = np.flip(B)
    for i in range(len(A)):
        sum = 0
        B2 = deque(B1)
        B2.rotate(i + 1)
        for j in range(len(A)):
             sum+=(A[j]*B2[j])
        zt.append(sum)
    return zt
```

```python
def frequency_domain(A , B):
    a = np.fft.fft(A)
    b = np.fft.fft(B)
    z = np.fft.ifft(a*b)
    return z
```

```python
#1
t1 = time.time()
z_t = time_domain(x,y)
plt.stem(z_t)
t2 = time.time()
```

```python
#2
z_t2 = time_domain(XX,YY)
plt.stem(z_t2)
```

```python
#3
z = frequency_domain(x,y)
plt.stem(z)
```

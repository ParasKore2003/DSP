```python
# Consider the following infinite non periodic discrete time signal
# x[n] = {0 n < 0
#         1 0<= n< a
#         0 n>= a}
# 1) Compute its discrete time fourier transform theoretically
# 2) Visualize the magnitude of X(e^(jw)) for one period of |X(e^(jw))| in 10000
#    points where a = 20
# 3) Generate a finite sequence x1[n] of length N = 30, sunch that x1[n] = x[n]
#    for n=0,1,.., N - 1. Compute DFT and plot its magnitude
#    try for N = 50, 100, 1000 and compare with plots in 2)
```

```python
#Libraries required for the experiment
import numpy as np
import matplotlib.pyplot as plt
```

```python
def plotter(n):
    W = [1 for i in range(0,20,1)] + [0 for j in range(20,n,1)]
    return np.fft.fft(W)
```

```python
a = 20
w = np.linspace(0,2 * np.pi,10000)
exponent = np.array([np.exp(complex(0,(-1*i))) - 1 for i in w ])
exponent1 = np.array([np.exp(complex(0,(-1*i*a))) - 1 for i in w ])
```

```python
y = ((exponent1)/(exponent))
plt.plot(abs(y))
```

```python
N1 = 30
X1 = plotter(N1)
plt.plot(abs(X1))
```

```python
N2 = 50
X2 = plotter(N2)
plt.plot(abs(X2))
```

```python
# Consider the signal x[n] = cos(2*pi*f0*n). Use the fft function to compute 
# and draw the DFT of the signal in N = 128 points, for f0 = 21/128 and f0 = 21/127
# explain the differences that we can see in the two sigal spectra
```

```python
def discrete(f,n):
    V = [np.cos(2 * np.pi * f * i) for i in range(0,n,1)]
    return np.fft.fft(V)
```

```python
f0 = 21/128
Y1 = discrete(f0,128)
plt.stem(Y1)
```

```python
f1 = 21/127
Y2 = discrete(f1,128)
plt.stem(abs(Y2))
```
